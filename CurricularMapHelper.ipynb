```python
import pandas as pd
```

```python
courses = pd.read_csv('data/course.csv', sep="\t")
schools = pd.read_csv('data/institution.csv', sep="\t")
```

```python
courses.set_index('course_id', inplace=True)
```

```python
schools.set_index('code', inplace=True)
```

```python
topics = pd.DataFrame(courses.course_topics.str.split(',').apply(pd.Series, 1).stack())
topics.index = topics.index.droplevel(1)
topics.columns = ['topic_name']
topics.topic_name = topics.topic_name.str.strip()
topics['topic_name_norm'] = topics.topic_name.str.lower()
```

```python
topics['course_title'] = courses.course_title
topics['institution_code'] = courses.institution_code
```

```python
topics.sort_values('topic_name')
```

```python
topics.to_csv('data/topics.csv', sep="\t")
```

```python
topickeys = topics.topic_name_norm.value_counts()
```

```python
topickeys.to_csv('data/topickey.csv', sep="\t")
```
