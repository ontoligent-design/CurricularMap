```python
import pandas as pd
import sqlite3
```

```python
db = sqlite3.connect('cmap2.db')
```

```python
courses_csv = 'data/course.csv'
courses = pd.read_csv(courses_csv, sep='\t')
courses.set_index('course_id', inplace=True)
```

```python
courses.head()
```

```python
topics = courses.course_topics.str.split(',').apply(pd.Series, 1).stack()
```

```python
topics = topics.apply(lambda x: x.strip())
```

```python
courses.to_sql('course', db, if_exists='replace', index=True)
```

```python
topics.to_sql('topic', db, if_exists='replace', index=True)
```
